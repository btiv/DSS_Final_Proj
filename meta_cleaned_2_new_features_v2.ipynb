```python
total_dict_key_author={}
total_dict_key_org={}
org_nation = {}
temp_dicts_key_author=[]
temp_dicts_key_org=[]
```

```python
def flatten_strip(list):
    # remove extra space around each element in the input list
    return [i.strip() for i in list]
```

```python
# Let me take one part of string in next block of main coding(that big "for" loop) as example.
# This method works but not elegant because I'm not good at regular expression.
# The first single unit we take care about begins with authors name in [ ] and organization name and nation name before a ";"
# ----------------------------------------------------------------first single unit-------------------------------------------------------------------
# '[Soltani, M.; Dehghani-Sanij, A. R.; Mahbaz, S. B.; Dusseault, M. B.] Univ Waterloo, Dept Earth & Environm Sci, Waterloo, ON, Canada;
#----------------------------------------------------------------------------------------------------------------------------------------------------
# so first I search "[" and "]" and take the string between them as a name assembly, then split this string by ";" 
# and strip extra space around each name. Now we have a list of name called "auth3"
# Second I search next ";" after a "]" and take the string between them as  a address assembly, then split this string by "," 
# and strip extra space around each address. The first word of address is organization name and the last word is nation name.

# The temp_dict_key_author is structured as {"author_1":[["org_1", "org_2"], X]}, X means how many times this author occurs
# in this paper's "address_and_author" part.

# The temp_dict_key_org is structured as {"org_1":[["author_1", "author_2"], X]}, X means how many times this organization occurs
# in this paper's "address_and_author" part.

# The total_dict_key_author and total_dict_key_org are in same format with two temp dictionary below, but it contains accumulated information
# in the loop

# The org_nation is structured as {"orgnization_1":"its_nation"}

# temp_dicts_key_author and temp_dicts_key_org are lists storing temp_dict_key_author and temp_dict_key_org respectively.

# Now for a certain paper, we can get the organization information if we focus on certain author, and we can get the author information
# if we focus on certain organization. Total information for each is also available.
```

```python
cleaned_nonan = cleaned.dropna(axis=0, subset=["address_and_author"])# delete the row with nan in this column
cleaned_nonan = cleaned_nonan.reset_index()
l = cleaned_nonan.shape[0]
l # number of rows after removing nan in column "address_and_author"
```

```python
# Use 100 instead of l to take an overview
for i in range(l):
    string = cleaned_nonan.loc[i,'address_and_author']

    state = -1
    
    temp_authors = ''
    temp_adds = ''
    
    temp_name = []
    temp_country = ''
    
    temp_dict_key_author={}
    temp_dict_key_org={}
    
    for s in string:
        if s == '[':
            state = 0
        elif s == ']':
            state = 1
        elif state == 1 and s == ';':
            state = 2
            
        if state == 0:
            temp_authors+=s
        elif state == 1:
            temp_adds+=s
        elif state == 2 and temp_authors:
            
            auth1 = temp_authors[1:]
            auth2 = auth1.split(';')
            auth3 = flatten_strip(auth2)
            
            add1 = temp_adds[2:]
            add2 = add1.split(',')
            add3 = flatten_strip(add2)
            
            org = add3[0]
            
#             nation = add3[-1]    # Use this line to make  'USA' as  ('NY 10598 USA', 'MA 02108 USA', 'TX 76019 USA', ...... )
            nation = add3[-1].split(' ')[-1] # Use this line to make ('NY 10598 USA', 'MA 02108 USA', 'TX 76019 USA', ...... )as 'USA'

            temp_authors,temp_adds='',''
            if org not in org_nation.keys():
                org_nation[org] = nation
                
            for author in auth3:
                if author in temp_dict_key_author.keys():
                    temp_dict_key_author[author][1]+=1
                    if org not in temp_dict_key_author[author][0]:
                        temp_dict_key_author[author][0].append(org)
                else:
                    temp_dict_key_author[author] = [[org],1]
                    
                if author in total_dict_key_author.keys():
                    total_dict_key_author[author][1]+=1
                    if org not in total_dict_key_author[author][0]:
                        total_dict_key_author[author][0].append(org)
                else:
                    total_dict_key_author[author] = [[org],1]
                    
            if org in temp_dict_key_org.keys():
                temp_dict_key_org[org][1]+=1
                for author in auth3:
                    if author not in temp_dict_key_org[org][0]:
                        temp_dict_key_org[org][0].append(author)
            else:
                temp_dict_key_org[org] = [auth3,1]
                
            if org in total_dict_key_org.keys():
                total_dict_key_org[org][1]+=1
                for author in auth3:
                    if author not in total_dict_key_org[org][0]:
                        total_dict_key_org[org][0].append(author)
            else:
                total_dict_key_org[org] = [auth3,1]
                
                
    temp_dicts_key_author.append(temp_dict_key_author)
    temp_dicts_key_org.append(temp_dict_key_org)
```
